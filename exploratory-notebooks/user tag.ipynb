```python
#jzw0203
# filter observations without null values in the column usertag
clk_0606_delna = clk_0606.dropna(axis = 0, subset = ['usertag'])
```

```python
# size of the dataframe
clk_0606_delna.shape
```

```python
# header
clk_0606_delna.head(5)
```

```python
# reset the index of filtered dataframe. Notice that in the previous dataframe the index skip some numbers, i.e. 4
clk_0606_delna = clk_0606_delna.reset_index(drop = True)
clk_0606_delna
```

```python
# usertag list
usertag_list = ["10006", "10024", "10031", "10048", "10052", "10057", "10059",
                  "10063", "10067", "10074", "10075", "10076", "10077", "10079",
                  "10083", "10093", "10102","10684", "11092","11278", "11379", 
                  "11423", "11512", "11576", "11632", "11680", "11724", "11944", 
                  "13042", "13403", "13496", "13678", "13776","13800", "13866", 
                  "13874", "14273", "16593", "16617", "16661", "16706", "16751", 
                  "10110","10111"]
```

```python
# have a dataframe containg all zeros
clk_0606_usertag = pd.DataFrame(np.zeros((clk_0606_delna.shape[0], len(usertag_list))), columns = usertag_list)
clk_0606_usertag.head(5)
```

```python
# add column bidid
clk_0606_usertag['bidid'] = clk_0606_delna.bidid
clk_0606_usertag.head(5)
```

```python
for i in range(clk_0606_delna.shape[0]):
    # extract usertag in row i
    temp_list = clk_0606_delna['usertag'][i]
    temp_split = temp_list.split(',')
    clk_0606_usertag.loc[i, temp_split] = 1
    
    
```

```python
clk_0606_usertag.head()
```

```python
# combine clk_0606
clk_0606_c = clk_0606_delna.join(clk_0606_usertag)
```
